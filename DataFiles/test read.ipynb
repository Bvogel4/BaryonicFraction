```python

#follow the main progenitor of a halo at redshift z=0, and plot the evolution of the mass of the main progenitor, plot the merger ratio as a function of redshift
history_df = elektra
import matplotlib.pyplot as plt

steps = sorted(history_df.keys(), reverse=True)
for halo in history_df[4096].keys():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    print(f'Halo {halo} at z=0')
    halo_0 = halo

    masses = []
    ratios = []

    for step in steps:
        if halo not in history_df[step]:
            masses.append(np.nan)
            ratios.append(np.nan)
            continue
        if history_df[step][halo]['merger_info']['has_merger']:
            print('\t',step,history_df[step][halo]['merger_info']['merger_ratio_stellar_only'])
            masses.append(history_df[step][halo]['final_state']['mvir'])
            ratios.append(history_df[step][halo]['merger_info']['merger_ratio_stellar_only'])
        else:
            masses.append(history_df[step][halo]['final_state']['mvir'])
            ratios.append(np.nan)

        halo = history_df[step][halo]['merger_info']['main_progenitor']

    ax[0].scatter(steps, masses, label=f'Halo {halo_0}')
    ax[1].scatter(steps, ratios, label=f'Halo {halo_0}')

    #plot major merger threshold
    ax[1].axhline(y=4, color='r', linestyle='--')
    #title
    fig.suptitle(f'Halo {halo_0}')

    ax[0].set_xlabel('Snapshot')
    ax[0].set_ylabel('Mvir')
    ax[0].set_title('Main Progenitor Mass Evolution')

    #log y scale
    ax[0].set_yscale('log')
    ax[1].set_yscale('log')
    #limit y axes for ratios
    ax[1].set_ylim(0.5, 100)
    ax[0].set_ylim(1e8, 1e11)

    ax[0].legend()

```

```python
#write a function to compare the 2 nested dicts {sim: {halo: {halo_data}}}
import pandas as pd
df = pd.DataFrame(columns=['sim', 'halo', 'time_old', 'ratio_old','n_old', 'time_new', 'ratio_new', 'n_new'])
def compare_dicts(new, old ):
    
    for sim in old.keys():
        #print(sim)
        #get halos from both dicts
        old_halos = list(old[sim].keys())
        new_halos = list(new[sim].keys())
        old_halos = [int(x) for x in old_halos]
        # combine the two lists and only keep unique values
        halos = list(set(old_halos + new_halos))
        #print(halos)
        
        
        for halo in halos:
            if str(halo) in old[sim].keys():
                time_old = old[sim][str(halo)]['times']
                ratio_old = old[sim][str(halo)]['ratios']
                n_old = len(time_old)
            else:
                time_old = []
                ratio_old = []
                n_old = 0
            if int(halo) in new[sim].keys():
                time_new = new[sim][int(halo)]['time']
                ratio_new = new[sim][int(halo)]['ratio']
                n_new = len(time_new)
            else:
                time_new = []
                ratio_new = []
                n_new = 0
            # add new row to dataframe
            #do not add if all lists are empty
            if time_old or ratio_old or time_new or ratio_new:
                df.loc[len(df)] = [sim, halo, time_old, ratio_old, n_old, time_new, ratio_new, n_new]
                
            
    return df
            
    
    
```
